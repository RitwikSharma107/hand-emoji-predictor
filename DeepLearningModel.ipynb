```python
from keras.layers import Dense, Flatten, Conv2D
from keras.layers import MaxPooling2D, Dropout
from keras.utils import np_utils, print_summary
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint
import pandas as pd
import numpy as np
import keras.backend as K
```

### Load Data

```python
df = pd.read_csv("Gestures_dataset.csv")
df.info()
```

### Modellling

```python
#Set up data
dataset = np.array(df) #Total columns = 50*50
np.random.shuffle(dataset) #Shuffle dataset

x = dataset
y = dataset
x = x[:, 1:2501] #All rows and all columns except first column
y = y[:, 0] #All rows and first column
```

```python
#Split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
#Normalization
x_train = x_train/255 
x_test = x_test/255 
```

```python
#Reshaping y_train and y_test
y_train = y_train.reshape(y_train.shape[0], 1)
y_train = y_train.T
y_test = y_test.reshape(y_test.shape[0], 1)
y_test = y_test.T
```

```python
print("No. of training samples: "+str(x_train.shape[0]))
print("No. of testing samples: "+str(x_test.shape[0]))
print("x_train shape: "+str(x_train.shape))
print("y_train shape: "+str(y_train.shape))     
print("x_test shape: "+str(x_test.shape))
print("y_test shape: "+str(y_test.shape))
```

```python
#Preprocessing data
image_x = 50
image_y = 50

train_y = np_utils.to_categorical(y_train)
test_y = np_utils.to_categorical(y_test)

train_y = train_y.reshape(train_y.shape[1],train_y.shape[2])
test_y = test_y.reshape(test_y.shape[1],test_y.shape[2])

x_train = x_train.reshape(x_train.shape[0], 50, 50, 1)
x_test = x_test.reshape(x_test.shape[0], 50, 50, 1)

print("x_train shape: "+str(x_train.shape))
print("x_test shape: "+str(x_test.shape))
print("train_y shape: "+str(train_y.shape))
print("test_y shape: "+str(test_y.shape))
```

```python
#Make Keras model

num_of_classes = 16 #Taken from train_y and test_y
    
model = Sequential() 
    
#2d Model layers
model.add(Conv2D(32, (5,5), input_shape=(image_x,image_y,1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2), padding='same'))
model.add(Conv2D(64, (5,5), activation='relu'))
model.add(MaxPooling2D(pool_size=(5,5), strides=(5,5), padding='same'))
    
model.add(Flatten()) #Converts to 1d model
#1d Model layers
model.add(Dense(1024, activation='relu'))
model.add(Dropout(0.6)) #Only 0.6 of random hidden layrs are used to prevent from overfitting 
model.add(Dense(num_of_classes, activation='softmax'))
```

```python
#Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
filepath = "hand_gestures.h5"
checkpoint1 = ModelCheckpoint(filepath, monitor='val_acc', verbose=0, save_best_only=True, mode='max')
callbacks_list = checkpoint1
```

```python
#Fit and evaluate the model

model.fit(x_train,train_y, validation_data=(x_test,test_y), epochs=1, batch_size=64, callbacks = [callbacks_list])

score = model.evaluate(x_test, test_y, verbose=0)
print("CNN Error: %.2f%%" % (100-score[1] * 100))
print_summary(model)
```

```python
#Save the model
model.save("Predictor_model.h5")
```
